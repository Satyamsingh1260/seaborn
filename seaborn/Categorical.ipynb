oxplot aur Violinplot 

Boxplot aur Violinplot dono graphs hote hain jo data ka distribution (data ka phelaav) dikhate hain.

Boxplot (ya box-and-whisker plot) quantitative data ko simple tareeke se dikhata hai taaki hum alag-alag variables ko compare kar sakein â€” jaise different categories ya groups.

Box data ke quartiles dikhata hai (25%, 50%, 75% values).

Whiskers box ke bahar tak jaate hain aur data ka baaki hissa dikhate hain.

Jo points bahut zyada door hote hain unhe outliers bolte hain.
Outliers ek special method se decide kiye jaate hain jo IQR (inter-quartile range) pe based hota hai.

Violinplot bhi distribution dikhata hai, par woh shape ke form mein dikhata hai taaki pata chale data kin values par zyada concentrated hai.<!--  -->

```python
sns.violinplot(x ='day',y= 'total_bill',data = df)
               
```

 1. Stripplot

Stripplot ek scatter plot jaisa hota hai, bas isme ek variable categorical hota hai (jaise: Male/Female, City types, etc.).

Har data point ko ek line/strip par plot kiya jata hai.

Ye khud bhi use ho sakta hai,
but mostly ise boxplot ya violinplot ke saath use karte hain
taaki har ek observation (data point) clearly dikh jaye.

2. Swarmplot

Swarmplot stripplot jaisa hi hota hai, lekin points overlap nahi hote.

Isme points ko thoda adjust kiya jata hai (sirf categorical axis par)
taaki koi point ek doosre ke upar na aaye.

Isse distribution clarity se dikhai deta hai.

Lekin bahut zyada data points ho toh thoda slow ho jata hai
aur sab poin<!--  -->

```python
sns.stripplot(x='day', y='total_bill', data=df)


```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

```python
sns.violinplot(x ='day',y= 'total_bill',data = df)
sns.swarmplot(x='day', y='total_bill', data=df)
```
