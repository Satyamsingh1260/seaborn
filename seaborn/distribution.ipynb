```python
#  we use distibution plots to visualize the distribution of quaNTIATIVE data
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

# DISTPLOT


```python
sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'],bins = 50)
# to increase the size
```

```python
sns.displot(df['total_bill'],bins = 30,kde = True)
```

```python
plt.subplot(1,2,1)
sns.displot(df['total_bill'],bins = 30,kde = True)
plt.subplot(1,2,2)
sns.displot(df['tip'],bins = 30,kde = True)
```

# JOINT PLOT

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

# pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue= 'sex')
```

```python
sns.pairplot(df,hue= 'sex',palette ='rainbow' )
```

```python
sns.pairplot(df,hue= 'size',palette ='rainbow' )
```

# rugplot

```python
sns.rugplot(df['total_bill'])
```

```python
sns.rugplot(df)
```
